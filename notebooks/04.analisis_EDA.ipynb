Histograma de fechas

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')

plt.figure(figsize=(10, 4))
plt.hist(df['date'], bins=60)
plt.title("Histogram of dates")
plt.xlabel("Date")
plt.ylabel("Number of records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Mapa de correlación

```python
df_mapa = df.copy()

df_mapa['y_1'] = df_mapa['y'].map({'yes': 1, 'no': 0})
df_mapa['y_1'] = df_mapa['y_1'].astype('Int64')

cols_numericas = [col for col in df_mapa.select_dtypes(include=['float64','Int64']).columns]
block = df_mapa[cols_numericas]
mapa = block.corr()


plt.figure(figsize=(10,8))
sns.heatmap(
    mapa,
    annot=True,
    fmt='.2f',
    vmin=-1, vmax=1,
    cmap='coolwarm',
    square=True,
    linewidths=.5,
    annot_kws={"fontsize": 6},
    cbar_kws={"shrink": 0.5} 
)
plt.title('Correlation map')
plt.tight_layout()
plt.show()
```

Gráficos de edades

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Gráfico 1: Histograma de edades ---
sns.histplot(df["age"], bins=30, ax=axes[0])
axes[0].set_title("Customer age distribution")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")

# --- Gráfico 2: Tasa de suscripción por edad ---

df["y_num"] = (df["y"] == "yes").astype(int)

subscription_rate = df.groupby("age")["y_num"].mean().reset_index()

axes[1].plot(
    subscription_rate["age"],
    subscription_rate["y_num"]
)
axes[1].set_ylim(0, 1)
axes[1].set_title("Subscription rate by age")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Subscription rate")

plt.tight_layout()
plt.show()
```

Gráficos de ingresos

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Gráfico 1: Histograma de ingresos ---
sns.histplot(df["income"], bins=30, ax=axes[0])
axes[0].set_title("Customer income distribution")
axes[0].set_xlabel("ncome")
axes[0].set_ylabel("Frequency")

# --- Gráfico 2: Tasa de suscripción por ingresos ---

df["income_bin"] = pd.qcut(df["income"], q=10)

subscription_rate_income = (
    df.groupby("income_bin", observed=True)["y_num"]
    .mean()
    .reset_index()
)

subscription_rate_income["income_mid"] = subscription_rate_income["income_bin"].apply(lambda x: x.mid)

axes[1].plot(
    subscription_rate_income["income_mid"],
    subscription_rate_income["y_num"],
    marker="o"
)

axes[1].set_ylim(0, 1)
axes[1].set_title("Subscription rate by income")
axes[1].set_xlabel("Income")
axes[1].set_ylabel("Subscription rate")
axes[1].grid(True, alpha=0.3)

plt.show()
```
