```python
#Tratamiento de datos
import pandas as pd
pd.set_option('display.max_columns', None)
#--------------

#Exploración de archivos
import sys
sys.path.append('../')
#--------------

#Funciones personalizadas
from src import sp_limpieza as sl
#--------------
```

```python
df_raw = pd.read_csv('../data/data_raw.csv')
df_raw.sample(5)
```

```python
df = df_raw.copy()
```

Cambio de nombres de columnas.

```python
#Columnas a minúsculas
df.columns = df.columns.str.lower()
#Cambio nombre columna 'numwebvisitsmonth'
df.columns = df.columns.str.replace('numwebvisitsmonth', 'num_web_visits_month')
```

dt_customer a date.

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format = '%Y-%m-%d')
```

Cambio de todos los valores a minus.

```python
sl.minus(df)
```

Crear columna de binario a str. Se utiliza para 'default', 'housing' y 'loan'.

```python
sl.binario_a_str(df,'default')
sl.binario_a_str(df,'housing')
sl.binario_a_str(df,'loan')
```

Cambio de ',' por '.' y conversión a float.

```python
#Se aplica a 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed'
sl.comas(df)
```

Cambio de columna 'date' a formato fecha.

```python
meses = {
    'enero': 'january',
    'febrero': 'february',
    'marzo': 'march',
    'abril': 'april',
    'mayo': 'may',
    'junio': 'june',
    'julio': 'july',
    'agosto': 'august',
    'septiembre': 'september',
    'octubre': 'october',
    'noviembre': 'november',
    'diciembre': 'december'
}

df['date'] = df['date'].replace(meses, regex=True)
df['date'] = pd.to_datetime(df['date'], format = '%d-%B-%Y')
```

Creación de columnas 'contact_month' y 'contact_year' que aparecian en el enunciado.

```python
df['contact_month'] = df['date'].dt.month
df['contact_year']  = df['date'].dt.year
```
