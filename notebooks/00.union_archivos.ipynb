```python
#Trataiento de datos.
import pandas as pd
pd.set_option('display.max_columns', None)
#--------------

#Exploración de archivos
import sys
sys.path.append('../')
#--------------

#Funciones personalizadas
from src import sp_union as su
#--------------
```

```python
#Importación de archivo csv.
bank_additional = pd.read_csv("../data/bank-additional.csv")

su.union_archivos(bank_additional)
```

```python
#Importación de todas las hojas de archivo excel.
customer_detail_lectura = pd.read_excel("../data/customer-details.xlsx", sheet_name=None)

#Unión de hojas de excel en un dataframe.
customer_detail = pd.concat(customer_detail_lectura.values(), ignore_index=True)

su.union_archivos(customer_detail)
```

```python
#Unión de ambas tablas por las columnas comunes "id_" y "ID".
data_raw = customer_detail.merge(bank_additional, left_on="ID", right_on="id_", how="inner")

su.union_archivos(data_raw)
```

```python
#Borrado de íncides de ambas tablas y columna "ID" sobrante.
data_raw.drop(columns=["Unnamed: 0_x", "Unnamed: 0_y","ID"], inplace=True)
data_raw.sample(5)
```

Al integrar ambos conjuntos de datos mediante un INNER JOIN utilizando el identificador único del cliente, se obtuvieron 43.000 registros. Los 170 clientes restantes presentes únicamente en el archivo de características demográficas no fueron incluidos en el análisis, ya que no participaron en la campaña de marketing y, por lo tanto, no disponían de información sobre el resultado de la oferta (y).

```python
#Exportación de dataframe
data_raw.to_csv('../data/data_raw.csv', index=False)
```
